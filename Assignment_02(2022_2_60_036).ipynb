#Outlier Detection

```python
# Boxplots for detecting outliers
sns.boxplot(df['Age'],color='orange')
plt.title('Boxplot for Age')
plt.show()

sns.boxplot(df['CGPA'],color='orange')
plt.title('Boxplot for CGPA')
plt.show()

sns.boxplot(df['Attendance_Percentage'],color='orange')
plt.title('Boxplot for Attendance_Percentage')
plt.show()
```

###Identifying CGPA Outliers Using Z-Score

```python
# Analyzing the boxplot: Outliers detected only in CGPA, also anomalies (CGPA greater than 4)
# Calculate Z-Scores
mean_cgpa = df['CGPA'].mean()
std_cgpa = df['CGPA'].std()
df['CGPA_Z'] = (df['CGPA'] - mean_cgpa) / std_cgpa

# determine Outliers
threshold = 3
outliers = df.loc[abs(df['CGPA_Z']) > threshold, ['CGPA', 'CGPA_Z']]

# print Outliers
print("Outliers Found in CGPA:")
print(outliers)

#total outliers
total_outliers = len(outliers)
print("Total Outliers:", total_outliers)
```

```python
# Histogram for CGPA before handling outliers
sns.histplot(df['CGPA'],color='purple',kde=True)
plt.title('CGPA Distribution Before outlier handling')
plt.show()
```

##Handling CGPA Outliers Using Capping Method

```python
# limit for CGPA
lower_limit = 2
upper_limit = 4

# Capping CGPA Outliers to Handle Extremes
df['CGPA'] = np.where(df['CGPA'] < lower_limit, lower_limit,np.where(df['CGPA'] > upper_limit, upper_limit, df['CGPA']))
print("Outliers capped successfully.")

#Boxplot after handling outliers
sns.boxplot(df['CGPA'],color='tan')
plt.title('Boxplot for CGPA')
plt.show()

# Histogram for CGPA after handling outliers
sns.histplot(df['CGPA'],color='olive', kde=True)
plt.title('CGPA Distribution After outlier handling')
plt.show()
```

#Feature Engineering

## Create a new feature called Engagement Score by combining Study Hours per Week and Attendance Percentage.

```python
# Calculate the engagement score (example: weighted average)
df['Engagement_Score'] = 0.6 * df['Study_Hours_per_Week'] + 0.4 * (df['Attendance_Percentage']/100)

# Display the updated DataFrame with the new feature
df.head()
```

```python
# Scatter plot between CGPA and Study Hours per Week
sns.scatterplot(data=df, x='Study_Hours_per_Week', y='CGPA', hue='Gender')
plt.title('CGPA vs Study Hours per Week')
plt.xlabel('Study Hours per Week')
plt.ylabel('CGPA')
plt.show()
# Visualize the new feature Engagement Score
sns.histplot(df['Engagement_Score'],color='brown', kde=True)
plt.title('Engagement_Score_Distribution')
plt.show()
```

## Featuring "Academic_Standing" that Categorizes CGPA Based on thresholds

```python
def categorize_cgpa(cgpa):
    if cgpa < 2.5:
        return 'Low'
    elif 2.5 <= cgpa <= 3.5:
        return 'Average'
    else:
        return 'High'

# Apply the function to create a new column
df['Academic_Standing'] = df['CGPA'].apply(categorize_cgpa)

# Display the updated DataFrame
df.head()
```

```python
# Visualize the new feature Academic Standing
sns.countplot(data=df, x='Academic_Standing', palette='viridis')
plt.title('Distribution of Academic Standing')
plt.xlabel('Academic Standing')
plt.ylabel('Count')
plt.show()
```

##Student Risk Level Based on Attendance, Job Status, and Satisfaction

```python
# Create Risk Level based on conditions
def risk_level(row):
    if row['Attendance_Percentage'] < 60 or row['Part_Time_Job'] == 'Yes' or row['Satisfaction_Level'] == 'Low':
        return 'High'
    return 'Low'
df['Risk_Level'] = df.apply(risk_level, axis=1)

# Display the updated DataFrame
df.head()

# Visualize Risk Level
sns.countplot(data=df, color='maroon',x='Risk_Level')
plt.title('Risk_Level_Distribution')
plt.show()
```
